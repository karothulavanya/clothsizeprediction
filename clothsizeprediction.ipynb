# Data manipulation

```python
#min and max are used to know the minimum and maximium values 
df = df.loc[(df['age'] >= 20.0) & (df['age'] <= 60)]
print('Min value: ', df['age'].min())
print('Max value: ',df['age'].max())
```

```python
#value_counts() is used for size column to know the how many values for each size
df['size'].value_counts()
```

```python
#here we used the mean() method to give the mean or average height and weight values to different size clothes
print('Average height for XXS clothes: ',df.loc[df['size'] == 'XXS']['height'].mean(), ' and average weight: ', df.loc[df['size'] == 'XXS']['weight'].mean())
print('Average height for S clothes: ',df.loc[df['size'] == 'S']['height'].mean(),' and average weight: ', df.loc[df['size'] == 'S']['weight'].mean())
print('Average height for M clothes: ',df.loc[df['size'] == 'M']['height'].mean(),' and average weight: ', df.loc[df['size'] == 'M']['weight'].mean())
print('Average height for L clothes: ',df.loc[df['size'] == 'L']['height'].mean(),' and average weight: ', df.loc[df['size'] == 'L']['weight'].mean())
print('Average height for XL clothes: ',df.loc[df['size'] == 'XL']['height'].mean(),' and average weight: ', df.loc[df['size'] == 'XL']['weight'].mean())
print('Average height for XXL clothes: ',df.loc[df['size'] == 'XXL']['height'].mean(),' and average weight: ', df.loc[df['size'] == 'XXL']['weight'].mean())
print('Average height for XXXL clothes: ',df.loc[df['size'] == 'XXXL']['height'].mean(),' and average weight: ', df.loc[df['size']=='XXXL']['weight'].mean())
```

```python
#As we already seen it in visualization XXL is not appearing much we are not taking these rows
df = df[df['size']!= 'XXL']
```

```python
#we are replacing the XXS and XXXL with XS and XXL as we didn't take XXL value rows we can replace it with this
#inplace=True means we are working with original dataset
df["size"].replace({"XXS": "XS", "XXXL": "XXL"}, inplace=True)
```

```python
#after not taking some rows and replacing them the value counts are
df["size"].value_counts()
```

```python
#after renaming and manipulating the dataset the scatter plot will be
#here we removed the unnecessary rows for a good prediction
plt.figure(figsize=(10,6), tight_layout=True)
ax = sns.scatterplot(data=df, x='height', y='weight',   hue='size', palette='Set2', s=60)
ax.set(xlabel='Height (cm)', ylabel='Weight (kg)')
ax.legend(title='Size', title_fontsize = 12) 
plt.show()
```

# Model building

```python
#here we splitted the data into training and testing
#it is very important to train the data before testing so it can give good results for prediction
#The prediction values are based on how we tain the data and testing matches with it or not
from sklearn.model_selection import train_test_split
X = df[['weight','age','height']]
y = df['size']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

# Decision Tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
print(accuracy_score(y_test, predictions)) 
```
